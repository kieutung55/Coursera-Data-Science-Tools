```python
def get_users():
    users=[]
    for user in RandomUser.generate_users(10):
        users.append({"Name":user.get_full_name(),"Gender":user.get_gender(),"City":user.get_city(),"State":user.get_state(),
                      "Email":user.get_email(),"DOB":user.get_dob(),"Picture":user.get_picture()})

    return pd.DataFrame(users)
```

## Another, more common way to use APIs, is through requests library.

```python
import requests
import json
```

```python
data = requests.get("https://web.archive.org/web/20240929211114/https://fruityvice.com/api/fruit/all")
```

Retrieve results using json.loads() function.

```python
results = json.loads(data.text)
results
```

Data is in JSON format, convert into a pandas dataframe

```python
pd.DataFrame(results)
```

The 'nutrition' column contains multiple subcolumns, so the data needs to be 'flattened' or normalized.

```python
df2 = pd.json_normalize(results)
df2
```

Extract some information from this dataframe.

```python
cherry = df2.query('name=="Cherry"').iloc[0]
family = cherry['family']
genus = cherry['genus']
(family,genus)
```

```python
banana= df2.query('name == "Banana"').iloc[0]
calories = banana['nutritions.calories']
calories
```
